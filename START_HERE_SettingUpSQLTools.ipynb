# SQLTools Installation, Connection, and Testing

## Installation

In order to work with Databases straight from VSCode, we need to install the extension.
1. Search for "SQLTools" by Matheus Teixeira in VSCode Exntensions Tab
    1. The extension will look like this:
    <img src="./visuals/sqltools_basic.png" style="width: 500px;">

    2. The Home page for the extension looks like this when you click on the extension:
    <img src="./visuals/sqltools_details.png" style="width: 500px;">

    3. If all else fails, please use this link: [SQLTools -  VSCode MarketPlace](https://marketplace.visualstudio.com/items?itemName=mtxr.sqltools) to go to the Visual Studio Marketplace online and follow the instructions to install the extension.
    4. NOTE: If you notice that the author that is in the screenshot (“mtxr”) and the one you see in VSCode (“Matheus Teixeira”) are different, do not panic as they are the same thing.
3. Go back to the Extensions tab
    1. Search for:
        1. “SQLTools SQLite” by Matheus Teixeira
        2. “SQLTools PostgreSQL/CockroachDB Driver” by Matheus Teixeira
    2. Install **both**
    3. Once both are installed, a popup will most likely appear asking for the extension to install other tools with Node
        1. Allow it
        2. Under the hood, these extensions use other Node modules to run these extensions.
        3. If no popup shows, restart VSCode. The popup may appear.
        4. If nothing shows even after a restart, you should be good to go.


## Connect a Database (SQLite)
1. Please Open the SQLTools Extension in VSCODE by clicking the database icon

    <img src="./visuals/databaseicon.png" style="width: 500px;">

2. Click the button that says “Add New Connection”  and your screen should look like this:

    <img src="./visuals/databasedriver.png" style="width: 500px;">

3. Click on the “SQLite (Node)”. You will then be greeted by this screen:

    <img src="./visuals/connectionname.png" style="width: 500px;">

4. Before entering any information, open your VSCode terminal and make a new SQLite file. You can use the command below:

    `touch test.sqlite`
   1. Alternatively, you can click the 'New File' button and name it `test.db`
  
Now we can start entering the information in the Connect Assistant, in the screenshot above.

5.  For the connection name, you can put “test” or another placeholder, as long as you’re mindful later on of what you’ve named your database.
6. Then for the Database file, select the “test.sqlite” you created.
    1. In case the “SELECT FILE” prompt doesn’t work, manually type in the path of where the SQLite file exists.
    2. You may also right-click the file and select ‘Copy Relative Path’ and paste it into the ‘Select File’ field
7. Then click on “TEST CONNECTION”
8. **If successful**:
    1. You should have a green alert saying “Successfully connected”.
    2. It should look like this:

        <img src="./visuals/successfulconnection.png" style="width: 500px;">

    3. You should also have a newly generated file called `test.session.sql`
        1. This is where we’ll run our SQL queries and use SQL as a language to communicate with our database

## Testing Your Connection

To test and make sure that our connection does actually work, do the following:

1. Copy the following query and paste it in your editor (`test.session.sql`):
    1. By doing so, we will create a new table with the following properties

       `CREATE TABLE Employees (`
       
            `EmployeeID INT PRIMARY KEY,`

            `FirstName VARCHAR(50),`

            `LastName VARCHAR(50),`

            `BirthDate DATE,`

            `HireDate DATE,`

            `Position VARCHAR(50),`

            `Salary DECIMAL(10, 2)`

       );
2. Then click on `Run on Active Connection` that is right above your cursor in the editor.

    
    
    1. Alternatively, you can open the VSCode Command Palette and type in “SQLTools” and then “Run this file”

3. We’ll see a few changes
    1. The Query History will reflect our actions / queries

    

    2. In the “Connections” window, right-click on “test” (or the name of your connection), and select ‘Refresh’.
        1. This should change to reflect the new table we’ve created with all its unique cells

    
        
    3. You may get a new tab open that appears blank. **Blank is *good*** because it simply means we’ve successfully run the connection with a well-executed query, as long as you can see the tables with their rows in 3.b under Connections. If instead the tab opened with an error, a step might have been completed incorrectly.

    

## Deleting Your Table
Now that you have successfully connected your database, you will need to return it to a blank template state by deleting your test tables.

1. Run the following query in your `test.session.sql` file

    `DROP TABLE Employees`

2. Use `Run Active Connection` to execute the query
3. Right-click on your database (`test`) on your SQLTools extension and click `Refresh` to ensure the table has been deleted


## Turning the Connection On/Off

1. To turn off: Right-click the connection name and select ‘Disconnect’
2. To turn on: Right-click the connection name and select ‘Connect’
