# Data Visualization with Pandas

This notebook demonstrates common plots used in data analysis:
- Line plot
- Bar chart
- Box plot
- Scatter plot
- Histogram

All examples use simple, synthetic data created with pandas.


# Demo 1 import pandas
Run the code block below to import pandas into this notebook environment. <br/>
- `import pandas as pd` -> organizes the data into tables
- `import numpy as np` -> generates or computes numeric data
- `import matplotlib.pyplot as plt` -> visualizes the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Make plots a bit nicer
plt.style.use("default")
%matplotlib inline
```

# Demo 2 Line plot
Let's create data for sales over time by running the code block below

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 170, 160, 180, 210]
```

In Code block 5, data was generated but using pandas convert the data into a table: <br>
- `pd.DataFrame(...)` creates a DataFrame (a table with rows and columns)
<br> The dictionary keys become column names and the dictionary values become the column data. The data above creates a table with:

- A "Month" column filled with values from months
- A "Sales" column filled with values from sales

Run code block 5 to create the table:

```python

df_line = pd.DataFrame({
    "Month": months,
    "Sales": sales
})

df_line
```

Now that the data has been organized into a table, create a line plot below where the `x-axis` will be the `Month` coloumn and the `y-axis` will be the `Sales` coloumn. Remeber to set the `kind="line"`.

```python
df_line.plot(
    x="Month",
    y="Sales",
    kind="line",
    marker="o",
    title="Monthly Sales Trend"
)

plt.ylabel("Sales ($)")
plt.show()
```

# Demo 3 Bar Chart — Category Comparison

Let's create data for sales over time by running the code block below. This time we are not storing our data in list but rather creating the data and table in one step.
- `pd.DataFrame(...)` creates a table (DataFrame)

- Each key in the dictionary becomes a column name

- Each list becomes the data for that column

```python
df_bar = pd.DataFrame({
    "Store": ["Store A", "Store B", "Store C", "Store D"],
    "Sales": [300, 450, 200, 500]
})

df_bar
```

Now that we have generated data, let's create a **bar chart** that will visualize sales by store. Create a **bar chart** below where the `x-axis` will be the `Store` coloumn and the `y-axis` will be the `Sales` coloumn. Remeber to set the `kind="bar"`. Run the code block below to generate the chart.

```python
df_bar.plot(
    x="Store",
    y="Sales",
    kind="bar",
    title="Sales by Store"
)

plt.ylabel("Sales ($)")
plt.show()
```

#Demo 4 - Box Plot
Let's use pandas to create a table of test scores. Let's breakdown the code block below <br>
`pd.DataFrame(...)` creates a table (DataFrame)

- The dictionary has one key, "Test Scores" setting the column name

- The list of numbers becomes the data in the "Test Scores", creating a single-column DataFrame with 10 rows.

```python
df_box = pd.DataFrame({
    "Test Scores": [65, 70, 72, 75, 78, 80, 85, 88, 90, 95]
})

df_box
```

Let's create a **box plot** with the "Test Scores" data above. In the code block below the method `df_box.plot(...)` creates the plot.
- kind="box" -> Tells pandas to create a box-and-whisker plot.
- title="Distribution of Test Scores" -> Adds a title at the top of the chart.

This boxplot visually summarizes:

- Minimum and maximum scores -> whiskers tails

- Median (middle score) -> represented by the green line

- Quartiles (spread of the data) - represented by the rectangle

- Potential outliers -> none on this plot, but would be represented as dots on the graphs.

```python
df_box.plot(
    kind="box",
    title="Distribution of Test Scores"
)

plt.ylabel("Score")
plt.show()
```

#Demo 5 Scatter Plot — Relationship Between Variables:
To create a realistic sample data for a scatter plot showing the relationship between hours studied and exam score, then displays the first few rows. Let’s go line by line on what the code block below is doing.
- `np.random.seed(0)` -> NumPy’s random number generator ensures the same random values every time you run the notebook
- `hours_studied = np.arange(1, 31)` -> creates numbers from 1 to 30 with each number representing the hours studied by one student
- `exam_score = 50 + (hours_studied * 1.5) + np.random.normal(0, 5, size=30)` -> Creating the y-values (Exam Scores)
Now that the data for the scatter plot has been generated. Use `pd.DataFrame(...)` to a table :
- Each row = one student

- Columns: "Hours Studied" → x-axis, "Exam Score" → y-axis


```python
np.random.seed(0)

hours_studied = np.arange(1, 31)  # 30 students
exam_score = 50 + (hours_studied * 1.5) + np.random.normal(0, 5, size=30)

df_scatter = pd.DataFrame({
    "Hours Studied": hours_studied,
    "Exam Score": exam_score
})

df_scatter.head()

```

Create the scatter plot using the `plot.()` method. Setting the `x-axis` to `Hours Studied` and the `y-axis` to `Exam Score`. To create the scatter plot set `kind="scatter"`. Run the code block below to generate the scatter plot.

```python
df_scatter.plot(
    x="Hours Studied",
    y="Exam Score",
    kind="scatter",
    title="Study Time vs Exam Score"
)

plt.show()
```

# Demo 6 Histogram — Frequency Distribution
Let's start by generating sample age data

- `np.random.seed(42)` -> Sets the random seed so the same random ages are generated every time.

- `np.random.randint(18, 65, size=50)` -> Generates 50 random integers between 18 and 64 (simulated ages).

- `pd.DataFrame({...})` -> Creates a pandas DataFrame with one column called **Age**.

- `df_hist.head()` -> Displays the first 5 rows of the DataFrame to preview the data.

Run the code block below to create the data :

```python
np.random.seed(42)

df_hist = pd.DataFrame({
    "Age": np.random.randint(18, 65, size=50)
})

df_hist.head()
```

Create the histogram plot using the `plot.()` method. Setting  `kind="hist"` and `bins=10`. Run the code block below to generate the histogram plot of age distribution.

```python
df_hist.plot(
    kind="hist",
    bins=10,
    title="Age Distribution"
)

plt.xlabel("Age")
plt.show()
```

# Demo 7 Pie Chart - Proportions
Let's create data for the **pie chart** by runing the code block below:

- `pd.DataFrame({...})` -> Creates a pandas DataFrame (table) from a dictionary.

- `"Company"` column -> contains the names of three companies.

- `"Market Share"` column -> contains each company’s market share values.

- `df_pie` -> Displays the DataFrame in a readable table format (in a Jupyter Notebook).

```python
df_pie = pd.DataFrame({
    "Company": ["Company A", "Company B", "Company C"],
    "Market Share": [45, 30, 25]
})

df_pie
```

Now that the data has been create, let's build a Pie chart by running the code block below:

- `df_pie.set_index("Company")`  -> sets **Company** as the index so each slice of the pie is labeled by company name.

- `.plot(kind="pie", y="Market Share")` -> creates a **pie chart** using the *Market Share* column.

- `autopct="%1.1f%%"` -> displays each slice’s percentage with one decimal place.

- `legend=False` -> removes the legend since labels appear directly on the pie.

- `title="Market Share Breakdown"` -> adds a title to the chart.

- `plt.ylabel("")` -> removes the default y-axis label for a cleaner look.

- `plt.show()` -> displays the pie chart.

```python
df_pie.set_index("Company").plot(
    kind="pie",
    y="Market Share",
    autopct="%1.1f%%",
    legend=False,
    title="Market Share Breakdown"
)

plt.ylabel("")
plt.show()
```
