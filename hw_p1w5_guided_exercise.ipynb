### Create a Score Keeping Function
33. In `seed.py`, add the following code to score your information:

````python
def compute_information_score(category: str, description: str, attributes: dict) -> int:
    score = 0
    if description and len(description) >= 80:
        score += 1
    if category == "electronics":
        if attributes.get("battery_life_hours") is not None:
            score += 1
    if category == "grocery":
        if attributes.get("is_gluten_free") is not None:
            score += 1
        if attributes.get("is_high_fiber") is not None:
            score += 1
    return score
````

This defines how “complete” each product’s information is and assigns a numeric score.

### Add data to your `product` table using Faker - Electronics
We specifically will add (or "seed") data into your products table using Faker.

34. Add the following code below your `computer_information_score` function declaration to add `electronics` to your `product` table

````python
def seed_electronics(session: Session, count: int) -> None:
    for _ in range(count):
        name = f"{fake.word().title()} Smartwatch"
        description = fake.text(max_nb_chars=120)
        battery_life = fake.random_element([None, fake.random_int(min=12, max=72)])
        attributes = {
            "battery_life_hours": battery_life,
        }
        product = Product(
            name=name,
            category="electronics",
            description=description,
        )
        product.set_attributes(attributes)
        product.information_score = compute_information_score(
            product.category,
            product.description,
            attributes,
        )
        session.add(product)
````

This generates fake smartwatch data — some complete, some incomplete — to test your scoring logic.

### Add data to your `product` table using Faker - Groceries
35. Add the following code below your `computer_information_score` function declaration to add `groceries` to your `product` table

````python
def seed_grocery(session: Session, count: int) -> None:
    for _ in range(count):
        name = fake.word().title()
        base_description = fake.sentence(nb_words=10)
        extra_detail = fake.random_element(
            [
                "",
                " Gluten-free.",
                " High in fiber.",
                " Gluten-free and high in fiber.",
            ]
        )
        description = (base_description + " " + extra_detail).strip()
        is_gluten_free = None
        is_high_fiber = None
        if "Gluten-free" in extra_detail:
            is_gluten_free = True
        if "High in fiber" in extra_detail or "high in fiber" in extra_detail:
            is_high_fiber = True
        attributes = {
            "is_gluten_free": is_gluten_free,
            "is_high_fiber": is_high_fiber,
        }
        product = Product(
            name=name,
            category="grocery",
            description=description,
        )
        product.set_attributes(attributes)
        product.information_score = compute_information_score(
            product.category,
            product.description,
            attributes,
        )
        session.add(product)
````

This creates `grocery` products, sometimes marking them as `gluten-free` or `high-fiber`, depending on the randomized description.

## Create a `main` function to execute the seeding to the database
36. Below the `seed_grocery` function declaration, add a `main` function to execute your seed functions

````python
def main() -> None:
    session: Session = SessionLocal()
    try:
        session.query(Product).delete()
        seed_electronics(session, count=10)
        seed_grocery(session, count=15)
        session.commit()
    finally:
        session.close()


if __name__ == "__main__":
    main()
````

Running this will clear existing products and insert 25 total items (10 electronics, 15 grocery).

## Create a database session helper

48. Directly beneath the `import` statements, paste the following function so the application has a consistent and reusable way to create database sessions whenever it needs to read from or write to the database. Keep in mind, we are only declaring the function, not yet calling it.

````python
def get_session() -> Session:
    return SessionLocal()
````


## Load products from the database

49. Below the `get_session()` function, add the following function `load_products`, which is responsible for querying the database, optionally filtering products by category, and safely closing the database session once the data has been retrieved. Keep in mind, we are only declaring the function, not yet calling it.

````python
def load_products(category_filter: str | None = None) -> list[Product]:
    session = get_session()
    try:
        query = session.query(Product)
        if category_filter and category_filter != "all":
            query = query.filter(Product.category == category_filter)
        products = query.order_by(Product.id).all()
        return products
    finally:
        session.close()
````

## Recalculate `information_score` for each individual product

50. Immediately after the `product-loading` helper, paste the following function which recalculates a product’s information score using the same logic defined in the seeding script, ensuring consistency across the application. Keep in mind, we are only declaring the function, not yet calling it.

````python
def compute_information_score_for_product(product: Product) -> int:
    attributes = product.get_attributes()
    score = 0
    if product.description and len(product.description) >= 80:
        score += 1
    if product.category == "electronics":
        if attributes.get("battery_life_hours") is not None:
            score += 1
    if product.category == "grocery":
        if attributes.get("is_gluten_free") is not None:
            score += 1
        if attributes.get("is_high_fiber") is not None:
            score += 1
    return score
````

## Add a helper to insert new products

51. Under the `information score` function, add the following helper which takes validated user input, builds the appropriate attributes dictionary based on the product category, calculates the information score, and commits the new product to the database. Keep in mind, we are only declaring the function, not yet calling it.

````python
def add_product(
    name: str,
    category: str,
    description: str,
    battery_life_hours: int | None,
    is_gluten_free: bool | None,
    is_high_fiber: bool | None,
) -> None:
    session = get_session()
    try:
        attributes = {}
        if category == "electronics":
            attributes["battery_life_hours"] = battery_life_hours
        if category == "grocery":
            attributes["is_gluten_free"] = is_gluten_free
            attributes["is_high_fiber"] = is_high_fiber
        product = Product(
            name=name,
            category=category,
            description=description,
        )
        product.set_attributes(attributes)
        product.information_score = compute_information_score_for_product(product)
        session.add(product)
        session.commit()
    finally:
        session.close()
````

## Convert products for display

52. After the `product insertion` helper, paste the following function, `product_to_dict`, which converts SQLAlchemy `Product` objects into plain dictionaries so they can be easily displayed inside Streamlit tables. Keep in mind, we are only declaring the function, not yet calling it.

````python
def product_to_dict(product: Product) -> dict:
    return {
        "id": product.id,
        "name": product.name,
        "category": product.category,
        "description": product.description,
        "attributes": product.get_attributes(),
        "information_score": product.information_score,
        "created_at": product.created_at,
        "updated_at": product.updated_at,
    }
````
